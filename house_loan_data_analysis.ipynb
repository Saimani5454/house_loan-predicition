# Task
Split the preprocessed data (`X_resample` and `y_resample`) into training and testing sets, then train a Logistic Regression model on the training data, evaluate its performance using relevant metrics, and finally save the trained model for later use in a Streamlit application.

## Split Data into Training and Test Sets

### Subtask:
Divide the preprocessed `X_resample` and `y_resample` data into training and testing sets to prepare for model training. This step will use `sklearn.model_selection.train_test_split`.


**Reasoning**:
To split the data into training and testing sets, I need to import the `train_test_split` function from `sklearn.model_selection`.



````python
from sklearn.model_selection import train_test_split
````

**Reasoning**:
Now that `train_test_split` is imported, I will split the `X_resample` and `y_resample` data into training and testing sets with a 20% test size, `random_state=42`, and `stratify=y_resample`.



````python
X_train, X_test, y_train, y_test = train_test_split(X_resample, y_resample, test_size=0.2, random_state=42, stratify=y_resample)
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
````

**Reasoning**:
Now that the data is split, I will import the Logistic Regression model from `sklearn.linear_model` to prepare for model training.



````python
from sklearn.linear_model import LogisticRegression
````

**Reasoning**:
With the `LogisticRegression` class imported, the next step is to instantiate the model and then fit it to the training data (`X_train`, `y_train`).



````python
model = LogisticRegression(random_state=42, solver='liblinear', max_iter=1000)
model.fit(X_train, y_train)
print("Logistic Regression model trained successfully.")
````

# Task
Evaluate the trained Logistic Regression model using `X_test` and `y_test` to calculate and display the accuracy, precision, recall, and F1-score.

## Evaluate the Model

### Subtask:
Evaluate the performance of the trained model using appropriate metrics such as accuracy, precision, recall, and F1-score on the test set. This will help understand how well the model generalizes.


**Reasoning**:
To evaluate the performance of the trained Logistic Regression model, I need to import the necessary metrics (accuracy, precision, recall, and F1-score) from `sklearn.metrics`.



````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
````

**Reasoning**:
Now that the evaluation metrics are imported, I will use the trained `model` to make predictions on the `X_test` data.



````python
y_pred = model.predict(X_test)
print("Predictions made on X_test.")
````

**Reasoning**:
Now that predictions have been made, I will calculate and print the accuracy, precision, recall, and F1-score using the `y_test` (actual values) and `y_pred` (predicted values).



````python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
````
